```python
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
```

# Problema 1

```python
def oscilador(y, t, b, k):
        x, v = y
        dydt = [v, -2*dseta*omega_n*v-np.power(omega_n,2)*x]
        return dydt

dseta = 0.0
omega_n = 3.0

y0 = [-1.0, 0.0]

t = np.linspace(0, 2*np.pi, 201)

y = odeint(oscilador, y0, t, args=(dseta, omega_n))
dseta = 0.1
y1 = odeint(oscilador, y0, t, args=(dseta, omega_n))
dseta = 0.2
y2 = odeint(oscilador, y0, t, args=(dseta, omega_n))
dseta = 0.4
y3 = odeint(oscilador, y0, t, args=(dseta, omega_n))
dseta = 0.7
y4 = odeint(oscilador, y0, t, args=(dseta, omega_n))
dseta = 1.0
y5 = odeint(oscilador, y0, t, args=(dseta, omega_n))
dseta = 2.0
y6 = odeint(oscilador, y0, t, args=(dseta, omega_n))

plt.figure(figsize=(10,5))
plt.plot(t, y[:, 0], 'b', label='dseta = 0')
plt.plot(t, y1[:, 0], 'g', label='dseta = 0.1')
plt.plot(t, y2[:, 0], 'r', label='dseta = 0.2')
plt.plot(t, y3[:, 0], 'y', label='dseta = 0.4')
plt.plot(t, y4[:, 0], 'black', label='dseta = 0.7')
plt.plot(t, y5[:, 0], 'purple', label='dseta = 1.0')
plt.plot(t, y6[:, 0], 'brown', label='dseta = 2.0')
plt.legend(loc='best')
plt.title('Solución del oscilador armónico amortiguado forzado con una fuerza de tipo sinoidal')
plt.xlabel('t')
plt.grid()
plt.show()
```
